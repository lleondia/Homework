```python
import pandas as pd
import numpy as np
import os


pathout = './data/'

if not os.path.exists(pathout):
    os.mkdir(pathout)
    
pathgraphs = './graphs/'
if not os.path.exists(pathgraphs):
    os.mkdir(pathgraphs)
```

## 1. Use the data from the Maddison Project to plot the evolution of total population across the world. 


```python
#Import databases from the Maddison Project
try:
    maddison_new = pd.read_stata(pathout + 'Maddison2018.dta')
    maddison_new_region = pd.read_stata(pathout + 'Maddison2018_region.dta')
    maddison_new_1990 = pd.read_stata(pathout + 'Maddison2018_1990.dta')
except:
    maddison_new = pd.read_stata('https://www.rug.nl/ggdc/historicaldevelopment/maddison/data/mpd2018.dta')
    maddison_new.to_stata(pathout + 'Maddison2018.dta', write_index=False, version=117)
    maddison_new_region = pd.read_stata('https://www.rug.nl/ggdc/historicaldevelopment/maddison/data/mpd2018_region_data.dta')
    maddison_new_region.to_stata(pathout + 'Maddison2018_region.dta', write_index=False, version=117)
    maddison_new_1990 = pd.read_stata('https://www.rug.nl/ggdc/historicaldevelopment/maddison/data/mpd2018_1990bm.dta')
    maddison_new_1990.to_stata(pathout + 'Maddison2018_1990.dta', write_index=False, version=117)
```

```python
maddison_new_region
```

```python
#Convert variable year to integer
basepop = maddison_new_region
basepop['year'] = basepop.year.astype(int)
basepop
```

```python
#Multiply population variable for a 1000, because this are the units of the variable
basepop['pop'] = basepop['pop'].apply(lambda x: x*1000)
basepop
```

```python
#Generate a new database only for the World, in order to get the total evolution of the population. 
basepopworld = basepop.loc[basepop.region_name=='World']
basepopworld = basepopworld.pivot_table(index='year',columns='region_name',values='pop',aggfunc='sum')
basepopworld
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
# Setup seaborn
sns.set()

```

```python
#Generate the plot from the evolution of the total population
plt.plot(basepopworld,"#34495e",label='World')
plt.xlabel('Years')
plt.ylabel('Population (in billions)')
plt.title('Evolution of total population across the world')
plt.show()
```

## 2. Plots the evolution of the share of the world population by countries and WB regions

### 2.1. Evolution of the share of the world population by country

```python
maddison_new
```

```python
#Merge maddison data by country with the one created previously that includes the date for World Aggregate
basepop_country=maddison_new.merge(basepopworld, on='year')
#Multiply population by a 1000
basepop_country['pop'] = basepop_country['pop'].apply(lambda x: x*1000)
basepop_country
```

```python
#Calculate the share of world population by country 
basepop_country['share']=(basepop_country['pop']/basepop_country['World'])*100
basepop_country
```

```python
#Pivot table to graph it later
basepop_country_F = basepop_country.pivot_table(index='year',columns='country',values='share',aggfunc='sum')
basepop_country_F
```

### ## 2.1. Evolution of the share of the world population by region

```python
from pandas_datareader import data, wb
countriesWB = wb.get_countries()
countriesWB['name'] = countriesWB.name.str.strip()
countriesWB
```

```python
#Change name of iso3c variable for countrycode in order to have the same variable name as the database basepop_country
countriesWB.rename(columns={'iso3c':'countrycode'}, inplace=True)
countriesWB
```

```python
#Merge the two databases and filter only the variables of interest
basepop_region = basepop_country[['countrycode', 'country', 'year', 'pop', 'World']].merge(countriesWB[['countrycode', 'region']], on=['countrycode'])
basepop_region  
```

```python
#Calculate the share of world population by country
basepop_region['share']=(basepop_region['pop']/basepop_region['World'])*100
basepop_region
```

```python

basepop_region_F = basepop_region.pivot_table(index='year',columns='region',values='share',aggfunc='sum')
basepop_region_F
```

```python
sns.set(rc={'figure.figsize':(11,8)})
sns.set_style("ticks")
# Plot
fig, ax = plt.subplots()
sns.lineplot(data=basepop_region_F , alpha=1, style='country', dashes=False, markers=True)
ax.tick_params(axis = 'both', which = 'major')
ax.tick_params(axis = 'both', which = 'minor')
ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.set_title("Share of world population by region", fontsize=18)
plt.savefig(pathgraphs + 'indpotential-Dev-NonDev.pdf', dpi=300, bbox_inches='tight')
```

Using this data analyze the convergence of life expectanty, mortality and fertility.

```python
#We will use the database from the World Bank previously named base_world
base_WB_p.rename(columns={'SP.DYN.TFRT.IN':'fertility'}, inplace=True)
base_WB_p.rename(columns={'SP.DYN.CDRT.IN':'mortality'}, inplace=True)
base_WB_p.rename(columns={'SP.DYN.LE00.IN':'life_exp'}, inplace=True)
base_WB_p
```

```python
base_WB_p1960 = base_WB_p.loc[base_WB_p.year==1960]
base_WB_p1960
```

```python
base_WB_p1970 = base_WB_p.loc[base_WB_p.year==1970]
base_WB_p1980 = base_WB_p.loc[base_WB_p.year==1980]
base_WB_p1990 = base_WB_p.loc[base_WB_p.year==1990]
base_WB_p2000 = base_WB_p.loc[base_WB_p.year==2000]
base_WB_p2010 = base_WB_p.loc[base_WB_p.year==2010]
base_WB_p2018 = base_WB_p.loc[base_WB_p.year==2018]
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.reset_orig()
sns.set_context("talk")
# Plot
fig, ax = plt.subplots()
sns.kdeplot(base_WB_p1960.fertility, ax=ax, shade=True, label='1960', linewidth=2)
sns.kdeplot(base_WB_p1970.fertility, ax=ax, shade=True, label='1970', linewidth=2)
sns.kdeplot(base_WB_p1980.fertility, ax=ax, shade=True, label='1980', linewidth=2)
sns.kdeplot(base_WB_p1990.fertility, ax=ax, shade=True, label='1990', linewidth=2)
sns.kdeplot(base_WB_p2000.fertility, ax=ax, shade=True, label='2000', linewidth=2)
sns.kdeplot(base_WB_p2018.fertility, ax=ax, shade=True, label='2018', linewidth=2)
ax.set_title("Fertility rate density", fontsize=18)
ax.set_xlabel('Fertility rate')
ax.set_ylabel('Density of Countries')
plt.savefig(pathgraphs + 'y1960-2000-densityfertility.pdf', dpi=300, bbox_inches='tight')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.reset_orig()
sns.set_context("talk")
# Plot
fig, ax = plt.subplots()
sns.kdeplot(base_WB_p1960.life_exp, ax=ax, shade=True, label='1960', linewidth=2)
sns.kdeplot(base_WB_p1970.life_exp, ax=ax, shade=True, label='1970', linewidth=2)
sns.kdeplot(base_WB_p1980.life_exp, ax=ax, shade=True, label='1980', linewidth=2)
sns.kdeplot(base_WB_p1990.life_exp, ax=ax, shade=True, label='1990', linewidth=2)
sns.kdeplot(base_WB_p2000.life_exp, ax=ax, shade=True, label='2000', linewidth=2)
sns.kdeplot(base_WB_p2018.life_exp, ax=ax, shade=True, label='2018', linewidth=2)
ax.set_title("Life expectancy density", fontsize=18)
ax.set_xlabel('Life expectancy')
ax.set_ylabel('Density of Countries')
plt.savefig(pathgraphs + 'y1960-2000-densitylifeexp.pdf', dpi=300, bbox_inches='tight')
```
